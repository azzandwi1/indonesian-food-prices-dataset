```python
import pandas as pd
```

```python
def transform(file_path):
	data = pd.read_excel(file_path)
	df = data.copy()
	df = df.drop(columns="No")
	df = df.rename_axis('index',axis=1)
	df = df.T.reset_index()
	df.columns = df.iloc[0]
	df = df.drop(df.index[0])
	
	col_names = ['tanggal',
             'beras','beras_kb1','beras_kb2','beras_km1','beras_km2','beras_ks1','beras_ks2', 
             'daging_ayam','daging_ayam_rs',
             'daging_sapi','daging_sapi_k1','daging_sapi_k2',
             'telur_ayam','telur_ayam_rs',
             'bawang_merah','bawang_merah_sedang',
             'bawang_putih','bawang_putih_sedang',
             'cabai_merah','cabai_merah_besar','cabai_merah_keriting',
             'cabai_rawit','cabai_rawit_hijau','cabai_rawit_merah',
             'minyak_goreng','minyak_goreng_curah','minyak_goreng_merk1','minyak_goreng_merk2',
             'gula_pasir','gula_pasir_premium','gula_pasir_lokal'
            ]
	df.columns = col_names
	
	return df
```

```python
df_2018 = transform("raw_data/2018.xlsx")
df_2019 = transform("raw_data/2019.xlsx")
df_2020 = transform("raw_data/2020.xlsx")
df_2021 = transform("raw_data/2021.xlsx")
df_2022 = transform("raw_data/2022.xlsx")
df_2023 = transform("raw_data/2023.xlsx")
df_2024 = transform("raw_data/2024.xlsx")
```

```python
frames = [df_2018, df_2019, df_2020, df_2021, df_2022, df_2023, df_2024]
daily = pd.concat(frames)
```
